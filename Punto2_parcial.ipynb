```python
import numpy as np # importa numpy
import matplotlib.pyplot as plt # importa matplotlib
from scipy.integrate import solve_ivp # importa scipy para resolver la ecuacion diferencial

def concentration_ode(t, C, k):# Definición de la ecuación diferencial
    return -k * C # ecuacion diferencial
C0 = 2 # concentracion inicial
k = 0.5 # constante de velocidad
t_span = [0, 10] # intervalo de tiempo en el que ocurre la reaccion
sol = solve_ivp(concentration_ode, t_span, [C0], args=(k,), t_eval=np.linspace(0, 10, 100)) # Solución de la ecuación diferencial
t = sol.t # solucion de la temperatura
C = sol.y[0] # solucion de la concentracion
C_5min = np.interp(5, t, C) # Concentración después de 5 minutos
print(f"Concentración después de 5 minutos: {C_5min:.4f} mol/L") # imprime el valor de la temperatura
plt.plot(t, C) # Gráfico de la concentración
plt.xlabel('Tiempo t (min)') # Titulo del grafico en x
plt.ylabel('Concentración C (mol/L)') # Titulo del grafico en y
plt.title('Concentración del Reactivo en el Reactor Batch') # Titulo del grafico
plt.grid(True) # Muestra la grilla del grafico
plt.show() # Muestra el grafico
```
